```python
# Sincronización con drive
folder_path = "/content/drive/My Drive/Data/Natalidad/Nacimientos_minsal/"

# Carga de archivos csv, con separación de columnas por ;
df_1992_2000 = pd.read_csv(folder_path + "Serie_Nacimientos_1992_2000.csv", encoding='latin1', sep=';')
df_2001_2019 = pd.read_csv(folder_path + "Serie_Nacimientos_2001_2019.csv", encoding='latin1', sep=';')
df_2020_2022 = pd.read_csv(folder_path + "Serie_Nacimientos_2020_2022.csv", encoding='latin1', sep=';')

# Limpiar nombres de columnas eliminando espacios y convertir todos los nombres a mayúsculas para evitar discrepancias
for df in [df_1992_2000, df_2001_2019, df_2020_2022]:
    df.columns = df.columns.str.strip().str.upper()

# Agregar columna de origen a cada DataFrame
df_1992_2000["PERIODO"] = "1992-2000"
df_2001_2019["PERIODO"] = "2001-2019"
df_2020_2022["PERIODO"] = "2020-2022"

# Obtener los conjuntos de columnas de cada DataFrame
cols_1992_2000 = set(df_1992_2000.columns)
cols_2001_2019 = set(df_2001_2019.columns)
cols_2020_2022 = set(df_2020_2022.columns)

df_nacimientos = pd.concat([
    df_1992_2000,
    df_2001_2019,
    df_2020_2022
], ignore_index=True, sort=False)

# Visualizar resumen de NaNs por columna agrupado por PERIODO
#nan_summary = df_nacimientos.groupby("PERIODO").apply(lambda x: x.isna().sum()).transpose()

# Mostrar resumen
import seaborn as sns
import matplotlib.pyplot as plt

#plt.figure(figsize=(15,10))
#sns.heatmap(nan_summary, annot=True, fmt=".0f", cmap="YlOrRd")
#plt.title("Cantidad de valores faltantes por columna y periodo")
#plt.show()
```
